```python
import pandas as pd

url_arquivo = 'youtube_dislike_dataset.csv'

dados = pd.read_csv(url_arquivo)

pd.DataFrame(dados)
```

```python
# Escolhi dois entre Likes e Dislikes para focar

df = pd.DataFrame(dados)

tabela_youtube = df[['likes', 'dislikes']]
tabela_youtube
```

```python
# 1 - Filtrando por Likes e Dislikes para comparar alto, médio e baixo

dados_youtube = (df['likes'] > 1000000) & (df['dislikes'] > 5000)
dados_resultados = df[dados_youtube].copy()

dados_resultados
```

```python
# 2 - Filtrando por soma total Likes e menor Dislikes

soma_total = dados_resultados['likes'].sum()
menor_dislikes = dados_resultados['dislikes'].min()

soma_total, menor_dislikes
```

```python
# 3 - Filtrando por menos likes para ver nome do Título

dados1 = dados_resultados.groupby('likes')
lista = dados1['title'].apply(list).min()
lista

```
